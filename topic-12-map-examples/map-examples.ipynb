```python
import folium

m = folium.Map(location=[41.1537, -81.3579], zoom_start=5)

# Add a circle marker
folium.Circle(
    location=[41.4993, -81.6944],
    radius=1000,  # Radius in meters
    color="blue",
    fill=True,
    fill_color="blue",
    fill_opacity=0.2,
).add_to(m)

folium.Marker(
        location=[41.1537, -81.3579],
        popup="Kent",
        icon=folium.Icon(icon="bell", color="blue"),  # Change icon color if needed
    ).add_to(m)

folium.Marker(
    location=[41.0814, -81.5190],
    icon=folium.DivIcon(
        html=f"""
        <div style="font-size: 16px; color: red; font-weight: bold; text-align: center;">
            Akron!
        </div>
    """
    ),
).add_to(m)

#  m.save("simple-kent-map.html")
m
```
